```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
# creat a column age_group
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_ frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import mysql.connector
import pandas as pd

# Replace with your MySQL server details
host = 'localhost'  # or your MySQL server IP
user = 'root'  # your MySQL username
password = 'Bronchitis823'  # your MySQL password
database = 'customer_behaviour'  # your database name

table_name = 'customer_shopping_behavior'  # the new table to create in MySQL

# Use the DataFrame already loaded earlier in the notebook
print('DataFrame rows:', len(df))
print('DataFrame columns:', list(df.columns))

# Map pandas dtypes to MySQL types

def mysql_type(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return 'BIGINT'
    if pd.api.types.is_float_dtype(dtype):
        return 'DOUBLE'
    if pd.api.types.is_bool_dtype(dtype):
        return 'TINYINT(1)'
    if pd.api.types.is_datetime64_any_dtype(dtype):
        return 'DATETIME'
    return 'TEXT'

col_defs = [f"`{col}` {mysql_type(dtype)}" for col, dtype in df.dtypes.items()]
create_sql = f"CREATE TABLE IF NOT EXISTS `{table_name}` ({', '.join(col_defs)});"

try:
    conn = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )
    print('Connected to MySQL database successfully!')

    cursor = conn.cursor()
    cursor.execute(create_sql)
    print(f'Created table if not exists: {table_name}')

    insert_sql = (
        f"INSERT INTO `{table_name}` ({', '.join(f'`{col}`' for col in df.columns)}) "
        f"VALUES ({', '.join(['%s'] * len(df.columns))})"
    )

    data_rows = [tuple(None if pd.isna(x) else x for x in row) for row in df.itertuples(index=False, name=None)]
    cursor.executemany(insert_sql, data_rows)
    conn.commit()
    print(f'Inserted {cursor.rowcount} rows into {table_name}')

    cursor.execute(f"SELECT COUNT(*) FROM `{table_name}`;")
    print('Table row count:', cursor.fetchone()[0])

    cursor.execute(f"SELECT * FROM `{table_name}` LIMIT 5;")
    for row in cursor.fetchall():
        print(row)

    cursor.close()
    conn.close()
    print('Connection closed.')

except mysql.connector.Error as err:
    print(f'Error: {err}')
```
